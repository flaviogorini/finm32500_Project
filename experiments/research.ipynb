```python
import pandas as pd
csv_path = f"nvda_alpaca_data.csv"

def read_csv(csv_path) -> pd.DataFrame:
    """ Reads and cleans price data from a CSV file. """
    
    price_data = pd.read_csv(csv_path)

    # set timestamp as index
    price_data = price_data.set_index('timestamp')

    # Basic cleaning
    price_data = price_data.dropna().drop_duplicates()
    price_data = price_data.sort_index()

    return price_data

nvda_data = read_csv(csv_path)
nvda_data.info()
nvda_data.head()
symbol = nvda_data['symbol'].iloc[0]
print(f"Symbol: {symbol}")
```

```python
import pandas_ta as ta

nvda_data["rsi_3"] = ta.rsi(nvda_data["close"], length=3)
nvda_data[["close", "rsi_3"]].tail(10)
```

```python
bb = ta.bbands(nvda_data["close"], length=20, lower_std=2.0, upper_std=2.0)
# get only BBP_20_2.0_2.0 column
print(bb['BBP_20_2.0_2.0'].iloc[-1])
bbpercent = bb['BBP_' + str(int(20)) + '_' + str(float(2.0)) + '_' + str(float(2.0))].iloc[-1]
print(bbpercent)
```

```python
# 20-period rolling z-score of close price
zscore = ta.zscore(nvda_data["close"], length=30)
zscore.iloc[-1]
```
