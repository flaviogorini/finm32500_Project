```python
# import NVDA data from yahoo finance
import yfinance as yf
import pandas as pd

# download data from 7 days ago to today with 1-minute intervals
nvda_data = yf.download(tickers='NVDA', period='7d', interval='1m')

# Display the first few rows of the data
print(nvda_data.head())

# Save data to a CSV file
nvda_data.to_csv('nvda_data.csv')
```

```python
#read data from csv and clean it
nvda_data = pd.read_csv('nvda_data.csv', index_col=0)
nvda_data.index = pd.to_datetime(nvda_data.index)
print(nvda_data.info())
print(nvda_data.head())
```
